```python
from pyspark.sql import SparkSession

# Tạo Spark session
spark = SparkSession.builder \
    .appName("read-parquet") \
    .master("spark://spark-master:7077") \
    .getOrCreate()

# Đọc file parquet
df = spark.read.parquet("hdfs://namenode:9000/data/raw/stock_data/year=2024")

# Hiển thị schema
print("\nSchema của dữ liệu:")
df.printSchema()

# Hiển thị số lượng records
print("\nSố lượng records:", df.count())

# Hiển thị vài dòng đầu tiên
print("\nDữ liệu mẫu:")
df.show(10, truncate=False)

# Hiển thị thống kê cơ bản
print("\nThống kê cơ bản:")
df.describe().show()
```
